```python
df['BasePay']=pd.to_numeric(df['BasePay'], errors = 'coerce')
average_basepay = df['BasePay'].mean()
average_basepay
```

```python
df['OvertimePay']=pd.to_numeric(df['BasePay'], errors = 'coerce')
highest_pay = df['OvertimePay'].max()
highest_pay
```

```python
job_title = df[df['EmployeeName'] == "JOSEPH DRISCOLL"]
job_title
```

```python
df['TotalPayBenefits'] = pd.to_numeric(df['TotalPayBenefits'], errors='coerce')
highest_paid = df[df['TotalPayBenefits'] == df['TotalPayBenefits'].max()]
highest_paid_person_name = highest_paid['EmployeeName'].values[0]
print(highest_paid_person_name)
print(highest_paid)
```

```python
df['TotalPayBenefits'] = pd.to_numeric(df['TotalPayBenefits'], errors='coerce')
lowest_paid = df[df['TotalPayBenefits'] == df['TotalPayBenefits'].min()]
lowest_paid_person_name = lowest_paid['EmployeeName'].values[0]
print(lowest_paid_person_name)
print(lowest_paid)
```

```python
filtered_data = df[(df['Year'] >= 2011) & (df['Year'] <= 2014)]
average_basepay_per_year = filtered_data.groupby('Year')['BasePay'].mean()
print(average_basepay_per_year)
```
